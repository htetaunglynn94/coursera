In this notebook, you will learn the process of Bagging  (Bootstrap Aggregation)  models for classification. Bagging is a method for generating multiple model versions and aggregating the ensemble of models to make a single prediction. For classification, aggregation performs majority vote when predicting a class. The various versions of the model are formed by performing Bootstrap sampling of the training set and using these to train each model in the ensemble .

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Give loops a progress bar
from tqdm import tqdm
# Data preprocessing functions
from sklearn import preprocessing
# Model accuracy
from sklearn import metrics as m
# tree functions
from sklearn import tree as t

from sklearn.model_selection import train_test_split as tts
from sklearn.tree import DecisionTreeClassifier as dtc
from sklearn.ensemble import BaggingClassifier as bc
from sklearn.model_selection import GridSearchCV as gscv
from sklearn.svm import SVC


# Suppress Numpy data type warnings
import warnings
warnings.filterwarnings('ignore', category=DeprecationWarning)

%matplotlib inline
````

This function calculates the train and test accuracy of a model


````python
def get_accuracy(X_train, X_test, y_train, y_test, model):
    train = model.predict(X_train)
    test  = model.predict(X_test)
    return {'test_accuracy': m.accuracy_score(y_test, test), 
            'train_accuracy': m.accuracy_score(y_train, train)}
````

This function creates visualizations of decision trees


````python
from  io import StringIO
# Used for decision tree visualization
import pydotplus
````

````python
# Plotting tree structure
def plot_tree(model, filename='tree.png'):
    '''
    This function is used to visualize and save the diagram as ".png" file and show the image.
    '''
    dot_data = StringIO() # to treat string like file object

    featureNames = [col for col in X_train.columns]
    out = t.export_graphviz(model,                           # decision tree estimator
                            feature_names = featureNames,    # feature names 
                            out_file = dot_data,             # output file name
                            class_names = ['left','stay'],   # each target classes' name
                            filled = True,                   # fill colours for identification
                            special_characters = True,       # use special characters (like <=)
                            rotate = False)                  # false='top-down', true='left-right'

    # DOT is a graph description format used for visualizing networks, trees, and graphs.
    graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
    # Save file as ".png"
    graph.write_png(filename)   
    # ---------------------------------------------------
    # Read image file and stored in 'img'
    img = mpimg.imread(filename)
    plt.figure(figsize=(100, 200))
    # Display image stored in 'img'
    plt.imshow(img, interpolation='nearest')
````

This function creates a graph of training accuracy vs how many estimators (Decision Trees) a BaggingClassifier uses


````python
def get_accuracy_bag(X, y, title, times=20, xlabel='Number Estimators'):
    '''
    This function works to visualize mean values of "Train Data Accuracy" and "Test Data Accuracy".
    1. Loops 20 times to split the data randomly. 
    2. Iterates number of estimators from 1 to 70.
    3. Creates accuracy data frame between training sets and testing sets.
    4. Plots both accuracy data in one image.
    '''
    #Iterate through different number of estimators and average out the results 
    n_estimators = [_ for _ in range(1, 71)]          # n_estimators: from 1 to 70
    train_acc = np.zeros((times, len(n_estimators)))  # 20 rows x 70 columns
    test_acc  = np.zeros((times, len(n_estimators)))  # 20 rows x 70 columns
    
    train_time = np.zeros((times, len(n_estimators)))  # 20 rows x 70 columns
    test_time  = np.zeros((times, len(n_estimators)))  # 20 rows x 70 columns
    
    # Average out the results
    for n in tqdm(range(times)):                       # loop 20 times
        X_train, X_test, y_train, y_test = tts(X, y, test_size=.3)
        for estimator in n_estimators:                 # loop 70 times
    
            # Iterate through different number of estimators and average out the results
            md = dtc(criterion='entropy', max_depth=10) # DecisionTreeClassifier moel
            bag = bc(estimator=md,                      # use DTC model
                     n_estimators=estimator,            # run DTC 30 times with different samples
                     bootstrap=True,                    # random sampling with replacement
                     random_state=0)                    # stable output
            bag.fit(X_train, y_train)                   # fit the model
    
            # Get accuracy scores
            accuracy = get_accuracy(X_train, X_test, y_train, y_test, bag)
    
            train_acc[n, estimator-1] = accuracy['train_accuracy']
            test_acc [n, estimator-1] = accuracy['test_accuracy']
            # --------------------------------------------------------
            # Visualization
            fig, ax1 = plt.subplots()
            ax2 = ax1.twinx()                         # use both values with common axis
    
            ax1.plot(train_acc.mean(axis=0), marker='o', c='b')   # plot train accuracy (blue color)
            ax2.plot(test_acc.mean(axis=0), marker='o', c='r')    # plot test accuracy (red color)
    
            ax1.set_xlabel(xlabel) 
            ax1.set_ylabel('Training accuracy', color='b')
            ax2.set_ylabel('Testing accuracy', color='r')
            plt.title(title + f"(n = {n+1} | n_estimators = {estimator})")
            plt.show()
            
    print(f"train_acc dimension --> {train_acc.shape} | test_acc dimension --> {test_acc.shape}")
````

## Bootstrap Sampling

Bootstrap Sampling is a method that involves drawing of sample data repeatedly with replacement from a data source to estimate a model parameter. Scikit-learn has methods for Bagging but its helpful to understand Bootstrap sampling.  We will import <code>resample</code>


````python
from sklearn.utils import resample
````

````python
churn_df[:5]
````

<h2 id="modeling">Decision  Tree Classifier  with Scikit-learn</h2>


A Decision  tree Classifier classifies a sample by learning simple decision rules inferred from the data. One problem with Decision  Tree Classifiers is overfitting; they do well with the training data, but they do not Generalize well. Trees have low bias and high variance; as such, they are a prime candidate for Bagging. Instability is another term used to describe models that overfit. Instability is characterized by a slight change in the training set that causes a drastic change in the model.  Let's show that Decision tree Classifiers are unstable.


We create and train a tree with a max depth of  5


````python
X_train, X_test, y_train, y_test = tts(X, y, test_size=.3, random_state=10)
tree_ = dtc(criterion='entropy', max_depth=5, random_state=10)
tree_.fit(X_train, y_train)
````

Now we can predict using our test set:


````python
yhat = tree_.predict(X_test)
yhat
````

We see the test error is much larger than the training error:


````python
get_accuracy(X_train, X_test, y_train, y_test, tree_)
````

We can plot the nodes of the tree:


````python
plot_tree(filename = "tree.png",model=tree_)
````

We can repeat the process but sampling different data points from the same dataset.  We see the tree still suffers from overfitting; in addition, the new tree is entirely different.


````python
X_train, X_test, y_train, y_test = tts(X, y, test_size=.3, random_state=5)
tree_ = dtc(criterion='entropy', max_depth=5, random_state=8)
tree_.fit(X_train, y_train)
print(get_accuracy(X_train, X_test, y_train, y_test, tree_))
plot_tree(filename='tree1.png', model=tree_)
````

<h2 id="evaluation">Bagging  for classification with  Scikit-learn</h2>
A Bagging classifier is an ensemble model that trains  base classifiers  on random subsets   of the original dataset  (Bootstrap Sampling by default), and then aggregate their individual predictions by voting.


Bagging improves models that suffer from overfitting; they do well on the training data, but they do not Generalize well. Decision Trees are a prime candidate for this reason, in addition, they are fast to train; We create a <code>BaggingClassifier</code> object,  with a Decision Tree as the <code>base_estimator</code>


````python
estimator = dtc(criterion='entropy', max_depth=4, random_state=2)
bag = bc(estimator = estimator,  # use DTC model
         n_estimators=30,        # run DTC 30 times with different samples
         random_state=0,         # stable output
         bootstrap=True)         # random sampling with replacement
````

We fit the model:


````python
bag.fit(X_train, y_train)
````

The method <code>predict</code>  aggregates the predictions by voting:


````python
bag.predict(X_test)
````

We see the training accuracy is slightly better but the test accuracy improves  dramatically:


````python
get_accuracy(X_train, X_test, y_train, y_test,  bag)
````

Here we can see the impact of adding more estimators (Decision Trees) using in Bagging on the testing and training accuracy


````python
get_accuracy_bag(X, y, "Customer Churn")
````

## Low Variance Example


Bagging does not improve result if the model has low Variance i.e. does reasonably well on the test and training data. Consider the SVM; the accuracy on the tests data and training data are similar


````python
clf = SVC(kernel='linear', gamma='scale')
clf.fit(X_train, y_train)
print(get_accuracy(X_train, X_test, y_train, y_test, clf))
````

Bagging the SVM does almost nothing:


````python
bagg = bc(estimator = clf,
          n_estimators = 10, 
          random_state = 0, 
          bootstrap = True)

bagg.fit(X_train, y_train)
print(get_accuracy(X_train, X_test, y_train, y_test, bagg))
````

````python
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ML0101EN-SkillsNetwork/labs/Module%203/data/cell_samples.csv")

df.head()
````

Now to determine the best parameters for `n_estimators` and the `max_depth` of the `base_estimator` we will use `GridSearchCV`


We can use GridSearch for Exhaustive search over specified parameter values. To  alter the base  model; we add the double underscore and the attribute value:

Here we are searching odd numbers from 1 to 39 for `n_estimators` and odd numbers from 1 to 20 for `max_depth` in the `base_estimator`


````python
[2 * n + 1 for n in range(20)]
````

````python
param_grid = {'n_estimators': [2 * n + 1 for n in range(20)], # for BaggingClassifier
              'estimator__max_depth': [2 * n + 1 for n in range(10)], # max_depth belongs to DecisionTreeClassifier
              'estimator__criterion': ['gini','entropy']} # criterion belongs to DecisionTreeClassifier
````

Create a `BaggingClassifier` object called `bag` with the `estimator` set to a `DecisionTreeClassifier` object where `random_state` = 0 and `bootstrap` = True


````python
estimator = dtc()
bag = bc(estimator = estimator, 
         random_state=0, 
         bootstrap=True)
````

</details>


Now we create a `GridSearchCV` object and search for the best parameters according to our `parameter_grid`


````python
gSearch = gscv(estimator = bag, 
               param_grid = param_grid, 
               scoring = 'accuracy', 
               cv = 3)
````

````python
gSearch.fit(X_train, y_train)
````

We can see the best accuracy score of the searched parameters was ~97%

And we can see the testing and training accuracy of the best estimator


````python
print(get_accuracy(X_train, X_test, y_train, y_test, gSearch.best_estimator_))
````
