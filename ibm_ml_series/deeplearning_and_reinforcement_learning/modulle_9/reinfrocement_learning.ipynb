## Reinforcement Learning Example

In this example from Reinforcement Learning, the task is to use tools from Machine Learning to predict how an agent should act. We will then use those predictions to drive the behavior of the agent. Ideally, our intelligent agent should get a much better score than a random agent.

## Key concepts:

- **Observation**: These are the states of the game. It describes where the agent currently is.
- **Action**: These are the moves that the agent makes.
- **Episode**: One full game played from beginning (`env.reset()`) to end (when `done == True`).
- **Step**: Part of a game that includes one action. The game transitions from one observation to the next.

__NOTE:__  
- `observation` $\to$ the new state after taking action
- `reward`      $\to$ immediate response from the environment  
  - `1` $\to$ reach the goal
  - `0` $\to$ all other cases (safe tile or hole)
- `terminated` $\to$ `True` if the episodes ended
  - Fall into a hole
  - Reach the goal
- `truncated` $\to$ rue if the episode ended due to external limits
  - Time limit reached
- `info` $\to$ Extra diagnostic data

````python
# we can put this process into a for-loop and see how the game progresses
current_observation, info = env.reset() # start a new game

for i in range(10): # run 5 moves

    new_action = env.action_space.sample() # same a new action

    observation, reward, terminated, truncated, info = env.step(new_action)
    done = terminated or truncated

    # here's a look at what we get back
    print(f"observation: {observation}, reward: {reward}, done: {done}, info: {info}")

    print(env.render())
````

````python
# Create frozen lake environment
env = gym.make('FrozenLake-v1')
# Plan to run 40,000 episodes
num_episodes = 40000
# Store ALL experiences from ALL episodes
life_memory = []

# Episode loop
for i in range(num_episodes):

    # Start a new episode
    old_observation, info = env.reset()
    # Controls episode end
    done = False
    # Total reward for this episode
    tot_reward = 0
    # Stores experiences only for this episode
    ep_memory = []

    # This loop runs until the agent falls into a hole or reaches the goal.
    while not done:
        # Choose a random move
        new_action = env.action_space.sample()
        # Apply the action
        observation, reward, terminated, truncated, info = env.step(new_action)
        done = terminated or truncated
        # tot_reward is usually 0 in Frozen Lake, Occasionally becomes 1 (goal reached)
        tot_reward += reward

        # Store experiences
        ep_memory.append({"observation": old_observation,
                          "action": new_action,
                          "reward": reward,
                          "episode": i})

        # Prepare for next state
        old_observation = observation

    # Number of steps in this episode
    num_steps = len(ep_memory)

    for i, ep_mem in enumerate(ep_memory):
        # Final outcome of episode (0 or 1)
        ep_mem["tot_reward"] = tot_reward
        # Small credit assigned earlier in episode
        ep_mem["decay_reward"] = i * tot_reward / num_steps

    # Appends this episode’s memory to all past memories
    life_memory.extend(ep_memory)

memory_df = pd.DataFrame(life_memory)
````
