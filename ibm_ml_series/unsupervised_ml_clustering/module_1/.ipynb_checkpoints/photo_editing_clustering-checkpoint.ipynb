```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans as KM

%matplotlib inline
```

```python
img = plt.imread('sky.jpg')
print(f"This photo is {img.shape[0]}px by {img.shape[1]}px.")

plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
img.shape
```

```python
def image_segmentation(image, n_clusters, method=1):
    X = image.reshape(-1, 3)
    km = KM(n_clusters=n_clusters).fit(X)
    seg = np.zeros(X.shape)

    for k in range(n_clusters):
        if method == 1:
            seg[km.labels_ == k] = km.cluster_centers_[k]
        elif method == 2:
            seg[km.labels_ == k] = 255 if km.cluster_centers_[k].mean() > 0.8 else 0
        else:
            print("This method haven't defined yet.")

    seg = seg.reshape(image.shape).astype(np.uint8)
    plt.imshow(seg)
    plt.axis('off')
    plt.show()
```

```python
image_segmentation(img, 2)
```

```python
image_segmentation(img, 2, 2)
```
