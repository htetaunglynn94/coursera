```python
x = np.array([[1,2], [3,4]], dtype=np.float64)
y = np.array([[5,6], [7,8]], dtype=np.float64)
```

```python
# Addidition
print(x + y)
print(np.add(x,y))
```

```python
# Subtraction
print(x - y)
print(np.subtract(x,y))
```

```python
# Multiplication
print(x * y)
print(np.multiply(x, y))
```

```python
# Division
print(x / y)
print(np.divide(x, y))
```

```python
# Square-root
print(np.sqrt(x))
```

```python
print(np.sum(x)) # Sum all elements in array
print(np.sum(x, axis=0)) # Sum of each column
print(np.sum(x, axis=1)) # Sum of each row
```

```python
print(np.mean(x)) # Mean of all elements in array
print(np.mean(x, axis=0)) # Mean of each column
print(np.mean(x, axis=1)) # Mean of each row
```

# Scipy

```python
from scipy import stats
import numpy as np
```

```python
# Create 10 normal random variables
print(stats.norm.rvs(size=10))
```

```python
from pylab import *

# Create some test data
dx = .01
X  = np.arange(-2,2,dx)
Y  = exp(-X**2)

# Normalize the data to a proper PDF
Y /= (dx*Y).sum()

# Compute the CDF
CY = np.cumsum(Y*dx)

# Plot both
plot(X,Y)
plot(X,CY,'r--')

show()
```

```python
# Compute the Normal CDF of certain values.
print(stats.norm.cdf(np.array([1,-1., 0, 1, 3, 4, -2, 6])))
```

```python
np.random.seed(1234)
# Generate 1000 students' T continuous random variables
x = stats.t.rvs(10, size=1000)
```

```python
print(x.min())
print(x.max())
print(x.mean())
print(x.var())
stats.describe(x)
```

# Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 3*pi, 0.1)
y = np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
x = np.arange(0, 3*np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.plot(x, y_sin)
plt.plot(x, y_cos)
plt.xlabel('x axis label')
plt.ylabel('y axis label')
plt.title('Sine and Cosine')
plt.legend(['Sine', 'Cosine'])
plt.show()
```

## Subplot

```python
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(2,1,1)
plt.plot(x, y_sin)
plt.title('Sine')

plt.subplot(2,1,2)
plt.plot(x, y_cos)
plt.title('Cosine')

plt.show()
```

# Seaborn

```python
import seaborn as sns
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
url = 'Cartwheeldata.csv'
df = pd.read_csv(url)
```

```python
df.head()
```

```python
# Scatter pot
sns.lmplot(x='Wingspan', y='CWDistance', data=df)
plt.show()
```

```python
# Scatterplot arguments
sns.lmplot(x='Wingspan', y='CWDistance', data=df,
         fit_reg=False,  # No regression line
         hue='Gender')   # Color by evolutioin stage
plt.show()
```

```python
# Construct Cartwheel distance plot
sns.swarmplot(x="Gender", y="CWDistance", data=df)
plt.show()
```

```python
# Boxplot
sns.boxplot(data=df.loc[:, ["Age", "Height", "Wingspan", "CWDistance", "Score"]])
plt.show()
```

```python
# Male Boxplot
sns.boxplot(data=df.loc[df['Gender'] == 'M', ["Age", "Height", "Wingspan", "CWDistance", "Score"]])
plt.show()
```

```python
# Female Boxplot
sns.boxplot(data=df.loc[df['Gender']=='F',['Age', 'Height', 'Wingspan', 'CWDistance']])
plt.show()
```

```python
# Male Boxplot
sns.boxplot(data=df.loc[df['Gender'] == 'M', ["Score"]])
plt.show()
```

```python
# Female Boxplot
sns.boxplot(data=df.loc[df['Gender'] == 'F', ["Score"]])
plt.show()
```

```python
# Histogram
sns.distplot(df.CWDistance)
plt.show()
```

```python
# Count plot
sns.countplot(x='Gender', data=df)
plt.xticks(rotation=45)
plt.show()
```
