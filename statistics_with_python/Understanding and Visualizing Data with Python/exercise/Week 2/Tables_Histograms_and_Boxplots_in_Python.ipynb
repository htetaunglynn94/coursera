```python
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
tips_data = sns.load_dataset('tips')
```

```python
tips_data.head()
```

```python
# Print out the summary statistics for quantitative variables
tips_data.describe()
```

```python
# Histogram of total_bill only
sns.distplot(tips_data['total_bill'], kde=False).set_title('Histogram of Total Bill')
plt.show()
# if "kde=True", figure will show line plot as well
```

```python
# Histogram of tips only
sns.distplot(tips_data['tip'], kde=False).set_title('Histogram of Total Tip')
plt.show()
```

```python
# Histogram of both total_bill and tip 
sns.distplot(tips_data['total_bill'], kde=False).set_title('Histogram of Total Bill')
sns.distplot(tips_data['tip'], kde=False).set_title('Histogram of Both Total B and Total Tiillp')
plt.show()
```

```python
# Boxplot of Total Bill
sns.boxplot(tips_data['total_bill']).set_title('Total bill boxplot')
plt.show()
```

```python
# Boxplot of Total Bill
sns.boxplot(tips_data['tip']).set_title('Total tip boxplot')
plt.show()
```

```python
# Boxplot of Total Bill and Total Tip
sns.boxplot(tips_data['total_bill'])
sns.boxplot(tips_data['tip']).set_title('Total bill and total tip boxplot')
plt.show()
```

```python
sns.boxplot(x = tips_data["tip"], y = tips_data["smoker"])
plt.show()
```

```python
# Create a boxplot and histogram of the tips grouped by smoker
sns.boxplot(x = tips_data["tip"], y = tips_data["smoker"])

g = sns.FacetGrid(tips_data, row = "smoker")
g = g.map(plt.hist, "tip")
plt.show()
```

```python
# Create a boxplot and histogram of the tips grouped by time of day
sns.boxplot(x = tips_data["tip"], y = tips_data["time"])

g = sns.FacetGrid(tips_data, row = "time")
g = g.map(plt.hist, "tip")
plt.show()
```

```python
# Create a boxplot and histogram of the tips grouped by the day of the week
sns.boxplot(x = tips_data["tip"], y = tips_data["day"])

g = sns.FacetGrid(tips_data, row = "day")
g = g.map(plt.hist, "tip")
plt.show()
```
