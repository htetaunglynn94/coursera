# Overfitting vs. Underfitting Explnation 

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

# Generate synthetic data with a quadratic pattern and higher noise
np.random.seed(0)
X = np.linspace(0, 10, 100).reshape(-1, 1)
y = (3 * X**2 + 2 * X + 1 + 50 * np.random.randn(100, 1)).flatten()  # Increased noise

plt.figure(figsize=(15,6))


# Plotting the data
plt.scatter(X, y, color='gray', label="Data")

# Underfitting - Linear Regression (too simple for quadratic data)
model_underfit = LinearRegression()
model_underfit.fit(X, y)
y_pred_underfit = model_underfit.predict(X)
plt.plot(X, y_pred_underfit, color='red', label="Underfitting (Linear)")

# Proper fit - Polynomial Regression (degree 2)
model_proper = make_pipeline(PolynomialFeatures(degree=2), LinearRegression())
model_proper.fit(X, y)
y_pred_proper = model_proper.predict(X)
plt.plot(X, y_pred_proper, color='green', label="Proper Fit (Degree 2)")

# Overfitting - Polynomial Regression (degree 15)
model_overfit = make_pipeline(PolynomialFeatures(degree=15), LinearRegression())
model_overfit.fit(X, y)
y_pred_overfit = model_overfit.predict(X)
plt.plot(X, y_pred_overfit, color='brown', label="Overfitting (Degree 15)")

# Finalize plot
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.title("Underfitting vs Proper Fit vs Overfitting")
plt.show()
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Generating a synthetic dataset
np.random.seed(0)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Splitting the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initializing and training the Linear Regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Predicting on the training set and calculating training accuracy (R-squared)
y_train_pred = model.predict(X_train)
training_accuracy = r2_score(y_train, y_train_pred)

# Predicting on the test set and calculating out-of-sample accuracy (R-squared)
y_test_pred = model.predict(X_test)
out_of_sample_accuracy = r2_score(y_test, y_test_pred)

# Printing the results
print(f"Training Accuracy (R-squared): {training_accuracy:.2f}")
print(f"Out-of-Sample Accuracy (R-squared): {out_of_sample_accuracy:.2f}")
```

`Training Accuracy` might be higher, as the model fits this data well.  
`Out-of-Sample Accuracy` might be slightly lower, indicating a more realistic performance on new data.
