Next, we just need to iterate through the `<th>` elements and apply the provided `extract_column_from_header()` to extract column name one by one


````python
def extract_column_from_header(row):
    """
    This function returns the landing status from the HTML table cell 
    Input: the  element of a table data cell extracts extra row
    """
    if (row.br):
        row.br.extract()      # remove line break
    if row.a:
        row.a.extract()       # remove hyperlink
    if row.sup:
        row.sup.extract()     # remove superscript
    
    # print("row.contents:", row.contents)
    column_name = ' '.join(row.contents) # white space is better for some cases
    # print("column_name:", column_name)
    
    # Filter the digit and empty names
    if not(column_name.strip().isdigit()):
        column_name = column_name.strip() # remove heading & trailing white spaces
        # print(column_name)
        return column_name    
````

**HTML Tags**  
`th` -> table header  
`td` -> table data (column)  
`tr` -> table row (row)  
`br` -> line break  
`a`  -> hyperlink  
`sup`-> superscript note

````python
column_names = []
# Apply find_all() function with `th` element on first_launch_table
# Iterate each th element and apply the provided extract_column_from_header() to get a column name
# Append the Non-empty column name (`if name is not None and len(name) > 0`) into a list called column_names

for row in first_launch_table.find_all('th'):
    name = extract_column_from_header(row)
    # print(name)
    if name is not None and len(name) > 0:
        column_names.append(name)
````

Next, we just need to fill up the `launch_dict` with launch records extracted from table rows.


Usually, HTML tables in Wiki pages are likely to contain unexpected annotations and other types of noises, such as reference links `B0004.1[8]`, missing values `N/A [e]`, inconsistent formatting, etc.


To simplify the parsing process, we have provided an incomplete code snippet below to help you to fill up the `launch_dict`. Please complete the following code snippet with TODOs or you can choose to write your own logic to parse all launch tables:


````python
len(BeautifulSoup.find_all(name = 'table', 
                           class_ = 'wikitable plainrowheaders collapsible'))
# BeautifulSoup.find_all(name = 'table', class_ = 'wikitable plainrowheaders collapsible')
# 'table' means all table in the HTML data
# 'wikitable plainrowheaders collapsible` are 3 types of classes
````

````python
def date_time(table_cells):
    """
    This function returns the data and time from the HTML  table cell
    Input: the  element of a table data cell extracts extra row
    
    Remark: table_cells.strings normally output the generator object to save
            memory and out put date and time data.
    """
    h = [data_time.strip() for data_time in list(table_cells.strings)][0:2]
    # print(h)
    return h
````

````python
def booster_version(table_cells):
    """
    This function returns the booster version from the HTML  table cell 
    Input: the  element of a table data cell extracts extra row
    """
    
    '''check list of all extracted string data'''
    # print([booster_version for i, booster_version in enumerate(row[1].strings)]) 
    
    '''check list of all extracted strin data at even places'''
    # print([booster_version for i, booster_version in enumerate(row[1].strings) if i % 2 == 0])
    
    
    # print(''.join([booster_version for i, booster_version in enumerate(row[1].strings) if i % 2 == 0][0:-1]))
            
    out = ''.join([booster_version for i,booster_version in enumerate( table_cells.strings) \
                   if i%2 == 0][0:-1])
    '''check the above result after removing last word and joining'''
    # print("joined words:", out)
    
    return out
````

````python
# Test why empty list return in 'booster_bersion'
a = ['H','T','E','T',' ','A','U','N','G']
b = ['H']
r1 = ''.join([name for i, name in enumerate(a) if i%2 == 0][0:-1])
r2 = ''.join([name for i, name in enumerate(b) if i%2 == 0][0:-1])
print('result 1:', r1)
print('result 2:', r2) # [0:-1] removes the last elemen happens emepty list []
````

````python
def get_mass(table_cells):
    # print(table_cells)
    # print("text:", table_cells.text)
    # print("string:", table_cells.string)
    # print(unicodedata.normalize('NFKD', table_cells.text).strip())
    
    mass = unicodedata.normalize("NFKD", table_cells.text).strip()
    
    '''Taking mass value from HTML table'''
    if mass:
        # print(mass.find("kg")) # return index position of "kg"
        # print(mass[0:mass.find('kg')+2])
        new_mass = mass[0:mass.find("kg")+2] # slice the text until found "kg"
    else:
        new_mass = 0
        
    return new_mass
````

````python
# Test how unicodedata.normalize() works
unicodedata.normalize('NFKD', "H₂O is water, x² + y² = r²")
````

````python
def landing_status(table_cells):
    """
    This function returns the landing status from the HTML table cell 
    Input: the  element of a table data cell extracts extra row
    """
    # print([i for i in table_cells.strings])
    out = [i for i in table_cells.strings][0].strip()
    # print(out)
    return out
````

````python
extracted_row = 0
for table_number, table in enumerate(BeautifulSoup.find_all('table', 'wikitable plainrowheaders collapsible')):
    
    # get table rows
    '''
    flag = False >> Header row
    flag = True  >> Data row
    '''
    for rows in table.find_all('tr'): # find data with 'tr'
        if rows.th: # if it is header row
            # print(rows.th)
            if rows.th.string: # if the header is string
                # print(rows.th.string.strip())
                flight_no = rows.th.string.strip() 
                flag = flight_no.isdigit() # if it's header, False
                # print(flight_no, ' : ', flag) # result >> Flight No.  :  False
        else: # if it is not header row
            flag = False # flag will be false
            
            
        '''
        The following 'row' data are stored like a list
        e.g, [ <td>D1<\td> , <td>D2<\td> , <td>D3<\td>, ... ]
        '''
        row = rows.find_all('td') # find data with 'td' under 'tr'
        # print("'td' values ******", row)

        
        if flag: # if it's not header (flag=True)
            launch_dict['Flight No.'].append(flight_no) # append 'flight_no' into dict
            # extracted_row += 1
            # print("row[0] ******", row[0])
            '''row[0] == <td>D1<\td> for instance'''
            # print(list(row[0].strings)[0:2]) # take the text only for data & time
            
            '''format date and assign into dict'''
            datetimelist = date_time(row[0])
            date = datetimelist[0].strip(',') # remove ',' from date
            # print(date)
            launch_dict['Date'].append(date)
            
            '''assign time into dict'''
            time = datetimelist[1] # since time format is correct, assign directly
            # print(time)
            launch_dict['Time'].append(time)
            
            '''get booster versioin name'''
            # print(row[1])
            bv = booster_version(row[1])
            # print("bv value:", bv)
            if not(bv): # if function returned empty list
                # print("not bv values:", row[1].a.string)
                bv = row[1].a.string
            # print(bv)
            launch_dict['Version Booster'].append(bv)
            
            '''assign launch_site into dict'''
            # print(row[2].a)
            launch_site = row[2].a.string
            # print(launch_site)
            launch_dict['Launch site'].append(launch_site)
            
            '''assign payload into dict'''
            # print(row[3].a)
            payload = row[3].a.string
            # print(payload)
            launch_dict['Payload'].append(payload)
            
            '''get payload mass'''
            # print(row[4])
            payload_mass = get_mass(row[4])
            # print("payload_mass:", payload_mass)
            launch_dict['Payload mass'].append(payload_mass)
            
            '''assign orbit data into dict'''
            # print(row[5])
            orbit = row[5].a.string
            # print(orbit)
            launch_dict['Orbit'].append(orbit)
            
            '''
            assign customer name into dict and avoid for AttributeError
            and IndexError
            '''
            try:
                # print("row[6]:", row[6])
                # print("a:", row[6].a)
                # print("string:", row[6].a.string)
                # print()
                customer = row[6].a.string
            # except (AttributeError, IndexError):
            except (AttributeError):
                # print("There is an AttributeError")
                customer = None
            launch_dict['Customer'].append(customer)
            
            '''assign outcome into dict'''
            # print(row[7])
            # print(row[7].string)
            # print(list(row[7].strings))
            launch_outcome = list(row[7].strings)[0].strip() # take first item from the list
            # print(launch_outcome)
            launch_dict['Launch outcome'].append(launch_outcome)
            
            '''assign landing status into dict'''
            booster_landing = landing_status(row[8])
            # print(booster_landing)
            launch_dict['Booster landing'].append(booster_landing)
````
